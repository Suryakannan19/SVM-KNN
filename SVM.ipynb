```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("mouse_viral_study.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x='Med_1_mL', y='Med_2_mL', hue='Virus Present', data=df)
x = np.linspace(0,10,100)
m = -1.1
b= 11
y = m*x + b
plt.plot(x,y, 'black')
```

```python
from sklearn.svm import SVC
```

```python
#help(SVC)
```

```python
X = df.drop('Virus Present',axis=1)
y = df['Virus Present']
```

```python
model = SVC(kernel='linear', C=1000) # C is inversly proportional and called the regularization parameter
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel = "linear", C=0.005)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y) #as C value is less more points are going near the margin
```

```python
model = SVC(kernel="rbf", C=1, gamma=2) #radial basis function ; non linear data func
model.fit(X,y) # dash line- margin and st line-hyperplane
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel = "sigmoid") ##Not recommended
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel = 'poly', C = 1, degree = 1)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1],'kernel' : ['linear', 'rbf']}
```

```python
grid = GridSearchCV(svm, param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.best_params_
```
